# LGBTI rights in Europe

# Introduzione

Negli ultimi decenni, si è discusso e sono stati riconosciuti sempre più diritti alle persone della comunità LGBTI. <br>
LGBTI è un termine ombrello che indica le persone che si indentificano come lesbiche, gay, bisessuali, transessuali, intersessuali. Il termine comprende quindi diversi orientamenti sessuali e l'identità di genere.

L'orientamento sessuale è l'attrazione emozionale, romantica e/o sessuale di una persona verso individui di sesso opposto, dello stesso sesso o di entrambi i sessi. [1]

L'identità di genere è come la persona si sente interiormente e il genere con cui si identifica. Questo può corrispondere al sesso assegnato alla nascita, oppure no. I giornali ancora raccontano le vite delle persone transessuali non usando i pronomi con cui quelle persone hanno scelto di identificarsi, insultandone quindi l'identità e addirittura chiamandoli con il nome registrato all'anagrafe, una violenza che si chiama deadnaming.
Intersessuali sono le persone che hanno un corpo che "non corrisponde alla definizione tipica dei corpi maschili o femminili" [2]. <br>
Nascono con un apparato riproduttivo o anatomia che non è definibile come maschile o femminile. A volte i dottori fanno operazioni su bambini intersessuali per poter assegnare un sesso che non è detto sia quello con cui si identificheranno una volta cresciuti. <br>
Molti intersessuali subiscono operazioni e trattamenti ormonali non voluti, mentre le persone transgender devono lottare per ottenerli.
Secondo gli studi di Anne Fausto-Sterling, l'1,7% della popolazione mondiale è intersessuale [3].

L'omotransfobia, transfobia e discriminazioni basate sull'orientamento sessuale hanno origine dalla suddivisione sociale dei ruoli. Chi non rispetta i ruoli imposti dalla società e sceglie di essere se stesso, viene discriminato. 

L'Unione Europea ha sviluppato degli standard di non discriminazione ed eguaglianza per persone LGBTI, ma questo sarà sufficiente ad evitare discriminazioni, aggressioni e molestie verso questa categoria marginalizzata? <br>
Le leggi sono sufficienti a modificare la società e a tutelare i cittadini LGBTI? <br>
La maggiore ricchezza di un paese porta a una maggiore consapevolezza e a una condizione migliore delle persone LGBTI? <br>
Questi sono i quesiti posti e le risposte sono ricercate in vari dataset presi in esame.

# Raccolta dati

Il primo dataset è fornito dalla <a href="https://fra.europa.eu/en/data-and-maps/2020/lgbti-survey-data-explorer">FRA</a>   (The European Union Agency for Fundamental Rights).
La FRA ha contribuito al progresso fornendo un'analisi degli aspetti legali e sociali riguardo i diritti fondamentali della comunità LGBTI in UE.
Nel 2019 la FRA ha condotto un sondaggio online per individuare se nell'Unione Europea le persone che fanno parte della comunità LGBTI (Lesbiche, Gay, Bisessuali, Transgender, Intersex) raggiungono pieno riconoscimento dei loro diritti fondamentali. Il sondaggio indaga sulla necessità di creare più leggi effettive e tutele per combattere la discriminazione, violenze e molestie con l'obiettivo di creare una società più giusta che tratta tutti in modo eguale.
Il sondaggio si interroga su come le persone LGBTI nell'UE (Unione Europea) sperimentano discriminazioni, molestie e violenze in diverse aree che fanno parte della vita.

La Commissione Europea ha richiesto alla FRA di raccogliere questi dati. La FRA ha quindi  creato un sondaggio online per raccogliere dati da tutti gli stati membri dell'UE (compreso il Regno Unito che faceva ancora parte dell'UE).
Chi ha risposto al sondaggio anonimo sono persone dai 15 anni in su, che si identificano come lesbiche, gay, bisessuali transgender o intersex. Il sondaggio online è stato condotto dal 27 maggio a 22 luglio del 2019, disponibile in tutte le lingue ufficiali dell'UE.

In totale 137508 persone hanno completato il sondaggio. La FRA ha lavorato in collaborazione con ILGA-Europe (European Region of the International Lesbian, Gay, Bisexual, Trans and Intersex Association).

Il dataset è strutturato nel seguente modo.
Sono presenti 7 campi:
- CountryCode: stato in cui è stato condotto il sondaggio
- question_code: identificativo univoco per ogni domanda
- target_group: categoria dell'intervistato ('Lesbian', 'Gay', 'Bisexual women', 'Bisexual men', 'Transgender', 'Intersex)
- question_lable: testo della domanda. Le domande prese in esame nell'analisi sono 6:
  1. Eviti di tenere la mano al partner del tuo stesso sesso in pubblico per paura di essere attaccato o minacciato?
  2. Negli ultimi 5 anni hai subito violenze fisiche o sessuali a causa dell'appartenenza alla comunità LGBTI?
  3. Hai denunciato alla polizia o a un'altra organizzazione l'ultima violenza fisica o sessuale causata dall'odio verso la comunità LGBTI?
  4. Hai subito violenze verbali negli ultimi 5 anni?
  5. Hai l'impressione che i pregiudizi e l'intolleranza contro le persone LGBTI siano incrementati o diminuiti negli ultimi 5 anni?
  6. Hai l'impressione che la violenza contro le persone LGBTI sia incrementata o diminuita negli ultimi 5 anni? 
 
- answer: testo della risposta 
- percentage: percentuale di intervistati 
- notes: commenti sulla dimensione del dataset

Associato a questo dataset, viene fornita un'altra tabella con indicato il numero di persone che hanno partecipato suddivise per categoria e paese.
Il dataset ha 8 campi:
- CountryID: codice identificativo del paese a cui si riferiscono i campi successivi

Sono presenti 14 campi:

- CSSSA LEGAL?: è l'incriminazione di consensuali atti sessuali tra persone dello stesso sesso (Criminalisation of Consensual Same-Sex Sexual Acts between Adults). I valori possono essere 'NO', 'YES', 'DE FACTO'.
- DATE OF DECRIM: anno della data di legalizzazione e 'NEVER CRIM' se non è ancora stato legalizzato il CSSSA.

- SAME SEX MARRIAGE: matrimonio tra persone dello stesso sesso ('YES','NO').
- CIVIL UNIONS: unioni civili ('YES','NO','LIMITED').
- JOINT ADOPTION: adozione congiunta ('YES','NO','LIMITED').
- SECOND PARENT ADOPTION: stepchild adoption ('YES','NO','LIMITED').

# Elaborazione dei dati

In questa fase i dati vengono sistemati per rendere i dataframe compatibili tra loro.

1. Il nome di colonna 'Total_n' di size_2019 viene sostituito con 'All' per essere compatibile con il target_group 'All' di 'domande'.

2. Nel dataset size_2019 vengono sostituiti i valori della colonna 'CountryID', che sono i codici identificativi dei paesi, con il nome degli stati usando il metodo map().
Se ad esempio il valore di colonna è 'AT', questo valore di chiave verrà sostituito dal suo valore che è 'Austria'.

3. I valori della colonna 'target_group' di 'domande' vengono resi compatibili con i valori della stessa colonna però di 'size_2019'.

4. Viene aggiunta nel dataframe domande la colonna con il numero di persone corrispondenti ai valori percentuali.
Viene creato un ciclo che itera su ogni riga di 'domande'. Dalla riga di cui si vuole calcolare il numero di persone si prende il nome del paese, il target group e con questi valori si estrae da size_2019 il numero che rappresenta il totale a cui applicheremo la percentuale. Con il totale e la percentuale si calcola il corrispondente numero di persone.
- Se il valore percentuale è ':', cioè non è presente, si mette il valore ':'. 
- Altrimenti si calcola il nuovo valore che viene inserito in una variabile X che costituirà la nuova colonna da inserire nel dataframe.
Dopo il completamento del ciclo viene aggiunta la Series X come colonna con lable 'number_people' nel dataframe domande.

5. Nel dataframe laws vengono mantenute solo le righe degli stati coinvolti nella nostra analisi ed eliminiamo le colonne non significative.

```python
#riscriviamo i nomi delle colonne per cambiare il nome 'Total_n', che raggruppa le categorie dello stesso paese, con 'ALL' 
#per essere compatibile con i nomi di categoria del dataframe 'domande'.
size_2019.columns = ['CountryID', 'Lesbian women', 'Gay men','Bisexual women','Bisexual men','Trans','Intersex','All']
size_2019.head()
```

```python
#mappiamo i codici identificativi degli stati con rispettivo nome dello stato.
#i valori originali della colonna 'CountryID' di size_2019 sono sostituiti con i nuovi valori indicati nel dizionario 
#che è il primo parametro nel metodo map(). 
#nella collezione si indica a destra la chiave (ID) e a sinistra il valore (nome del paese).

#copiamo il dataframe con gli ID dei paesi per usarlo successivamente
size_sigle=size_2019.copy(deep=True) #viene fatta una deep copy così i due dataframe non condividono la stessa area di memoria

size_2019["CountryID"] = size_2019["CountryID"].map( 
    {
        "AT": "Austria",
        "BE": "Belgium",
        "BG": "Bulgaria",
        "CY": "Cyprus",
        "CZ": "Czechia",
        "DE": "Germany",
        "DK": "Denmark",
        "EE": "Estonia",
        "EL": "Greece",
        "ES": "Spain",
        "FI": "Finland",
        "FR": "France",
        "HR": "Croatia",
        "HU": "Hungary",
        "IE": "Ireland",
        "IT": "Italy",
        "LT": "Lithuania",
        "LU": "Luxembourg",
        "LV": "Latvia",
        "MT": "Malta",
        "NL": "Netherlands",
        "PL": "Poland",
        "PT": "Portugal",
        "RO": "Romania",
        "SE": "Sweden",
        "SI": "Slovenia",
        "SK": "Slovakia",
        "UK": "United Kingdom",
        "EU-28": "EU-28",
    },
    na_action=None,
)
size_2019.head()
```

```python
#mappiamo i valori della colonna target_group in domande con corrispettivi valori di target_group di size_2019 

domande["target_group"] = domande["target_group"].map(
    {
        "All": "All",
        "Intersex people": "Intersex",
        "Trans people": "Trans",
        "Gay men": "Gay men",
        "Lesbian women": "Lesbian women",
        "Bisexual women": "Bisexual women",
        "Bisexual men": "Bisexual men"
    },
    na_action=None,
)
domande.head()
```

```python
#viene aggiunta in domande la colonna con il numero di persone corrispondenti alla percentuale

X=[] #viene creata la Series X dove verrà inserito il numero di persone per ogni riga di 'domande'
i=0 #viene inizializzato i a 0
for country in domande['CountryCode']: #ciclo sugli elementi della colonna 'CountryCode' in modo da iterare su tutte le righe del dataframe
 #nella variabile target_group viene memorizzato il valore di 'target group' della riga i di 'domande'. 
 target_group=domande['target_group'].iloc[i]  
 riga_corretta=size_2019.loc[size_2019['CountryID']==country] #memorizzata la riga con le dimensioni che ci occorrono
 percent=domande['percentage'].iloc[i] #valore percentuale di cui vogliamo calcolare il corrispondente numero di persone
 totale=riga_corretta[target_group].astype(int) #il numero totale di persone del paese 'country'
 #if (percent==np.NaN):
 if (percent==':'):
    #X.append(np.NaN)
    X.append(':')
 else:
    percent=int(percent) #trasformazione da stringa a int
    valore_new=(percent*totale)//100 #calcolo numero di persone corrispondenti a 'percent'
    valore_new=valore_new.to_string(index=False) #trasformazione in stringa
    X.append(valore_new) #aggiungiamo il valore alla Series X
 i=i+1
domande.insert(5,"number_people",X,True) #inseriamo la colonna nel dataframe 'domande'
domande.head()
```

```python
stati=domande['CountryCode'] #prendiamo i paesi dove è stato effettuato il sondaggio
laws=laws[laws['COUNTRY'].isin(stati)]#teniamo solo le righe che si riferscono agli stati selezionati
laws=laws.drop('MAX PENALTY',1) #eliminiamo la colonna perchè ha solo valori Nan
laws.head()
```

```python
INCR_prejudice=prej_country.loc[prej_country['answer']=='Increased a lot'] #prendiamo dal dataset le righe con la risposta 'Increased a lot'
INCR_prejudice=INCR_prejudice.loc[INCR_prejudice['CountryCode']!='EU-28']  #teniamo solo gli stati e non la media sull'UE
INCR_prejudice=INCR_prejudice.loc[INCR_prejudice['CountryCode']!='Poland'] #eliminiamo gli outlier
INCR_prejudice=INCR_prejudice.loc[INCR_prejudice['CountryCode']!='France'] 


# facendo un marge, al dataframe YES_attack aggiungiamo il PIL del paese di riferimento
INCR_prejudice=INCR_prejudice.merge(pil, left_on='CountryCode', right_on='CountryCode')
INCR_prejudice=INCR_prejudice.sort_values(by=['percentage']) #ordiniamo i valori percentuali

x=INCR_prejudice[' PIL'] #assegniamo a x i PIL dei vari paesi che saranno sull'asse x
y=INCR_prejudice['percentage'] #assegniamo a y le percentuali dei vari paesi che saranno sull'asse y

plt.scatter(x, y)  #viene creato lo scatter plot
plt.xlabel('PIL') #x label
plt.ylabel('Percentage') #y label

#visualizzazione della best line fit
#il metodo polyfit() stima i parametri della retta usando i dati, poly1d() crea un'equazione usando i coefficienti.
linear_model=np.polyfit(x,y,1) #viene creata un'equazione lineare, perchè viene passato 1 come terzo argomento di polyfit()
linear_model_fn=np.poly1d(linear_model)
x_s=np.arange(0,120)
plt.plot(x_s,linear_model_fn(x_s),color="green") #viene plottata la retta

plt.show()

#calcolo la correlazione e p-value
corr, p = sci.pearsonr(x, y)
print('Pearsons correlation: %.3f ' % corr)
print('p-value: %.3f' % p)
```

La correlazione è aumentata e può considerarsi moderata. p-value ora è sotto il valore soglia, quindi il valore di correlazione è statisticamente rilevante
